## Crawling wiki table
A note on how to crawl wiki table.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
wiki_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table_id = 'constituents'
```

```python
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
snp500_table = soup.find('table', attrs={'id': table_id})
print(snp500_table)
```

```python
df = pd.read_html(str(snp500_table))[0]
print(df)
```

```python
sorted_list = df.sort_values(by = ['GICS Sector', 'GICS Sub-Industry'])
sectors = sorted_list[['GICS Sector', 'GICS Sub-Industry', 'CIK']]
print(sectors)
print(set(sectors["GICS Sector"]))
sectors[sectors["GICS Sector"] == "Communication Services"]
```

```python
df.to_csv("./sp500.csv", index = False, header = True)
```
