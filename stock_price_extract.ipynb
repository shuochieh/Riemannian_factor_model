```python
import yfinance as yf
import pandas as pd
import json
```

```python
sp500comp = pd.read_csv("./sp500.csv")
sp500cik = sp500comp["CIK"]

file_name = "company_tickers.json"
cik_to_tickers_dict = {}

with open(file_name, "r") as f:
    file = json.load(f)
    for company in file.values():
        if cik_to_tickers_dict.get(str(company["cik_str"]), None):
            cik_to_tickers_dict[str(company["cik_str"])].append(company["ticker"])
        else:
            cik_to_tickers_dict[str(company["cik_str"])] = [company["ticker"]]
```

```python
save_for_later = []
res = {}
for cik in sp500cik:
    temp = sp500comp.loc[sp500comp["CIK"] == cik, "Symbol"]
    if len(temp) > 1:
        save_for_later.append(cik)
        continue
    if temp.item() in cik_to_tickers_dict[str(cik)]:
        res[cik] = temp.item()
    else:
        save_for_later.append(cik)
        
print(save_for_later)
```

```python
cik_to_tickers_dict = res
```

```python
tickers = list(cik_to_tickers_dict.values())
```

```python
data = yf.download(tickers, start="2000-01-01", end="2024-12-31")
```
